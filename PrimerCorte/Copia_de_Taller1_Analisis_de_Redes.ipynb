1. Cuales fueron las ventas totales?

```python
ventas_totales = df["Total Sales"].sum()
print("Las ventas totales fueron: $", round(ventas_totales, 2))
```

1.1 Cuanto fue la ganancia operacional total?

```python
ganancia_operacional_total = df["Operating Profit"].sum()
print("La ganancia operacional total fue: $", round(ganancia_operacional_total, 2))
```

1.2 Cuanto fue el margen promedio total ?

```python
margen_promedio = df["Operating Margin"].mean() * 100
print("El margen operativo promedio fue:", round(margen_promedio, 2), "%")
```

**1.3 Por que el margen promedio y no el total. Explique con palabras**



El **margen operativo** es un indicador expresado en forma de **proporci√≥n o porcentaje**.  
A diferencia de las ventas o de la ganancia, que son montos de dinero y por eso s√≠ se pueden sumar, el margen muestra la **relaci√≥n entre la ganancia y las ventas** de cada transacci√≥n.  

Si sum√°ramos todos los m√°rgenes fila por fila, el resultado no tendr√≠a sentido, porque estar√≠amos acumulando porcentajes de distintas operaciones.  
Por eso, lo correcto es calcular el **promedio del margen**, que nos da un valor representativo de la eficiencia general de la empresa.  

En este caso, el margen promedio fue de aproximadamente **42%**, lo que significa que, en promedio, de cada d√≥lar vendido, 42 centavos se convierten en ganancia operativa.



2. Cuales fueron las ventas totales por Retailer?

```python
ventas_por_retailer = df.groupby("Retailer")["Total Sales"].sum().sort_values(ascending=False)
print("Ventas totales por Retailer:")
print(ventas_por_retailer)
```

2.1 Cual fue la ganancia operacional total por retailer?

```python
ganancia_por_retailer = df.groupby("Retailer")[["Operating Profit"]].sum()
ganancia_por_retailer = ganancia_por_retailer.sort_values(by="Operating Profit", ascending=False)
print("Ganancia operacional total por Retailer:")
print(ganancia_por_retailer)
```

2.3 Tomando los dos resultados de arriba: El ganancia operacional total y las ventas totales divida ambos resultados. Le tiene que dar el margen total de cada Retailer.

```python
ventas_totales = df.groupby("Retailer")["Total Sales"].sum()
ganancia_total = df.groupby("Retailer")["Operating Profit"].sum()
margen_total = ganancia_total / ventas_totales
margen_total = margen_total * 100
print("üìä Margen Operacional Total por Retailer (%):")
print(margen_total.round(2))
```

2.4 Cual fue el margen operacional promedio por retailer?

```python
margen_promedio = df.groupby("Retailer")["Operating Margin"].mean() * 100
print("üìä Margen Operacional Promedio por Retailer (%):")
print(margen_promedio.round(2))
```

2.5 Por que los dos resultados anteriores le dieron diferente? Argumente

La diferencia se debe a la forma en que se calcula cada indicador:

- **Margen operacional total**: se obtiene dividiendo la **ganancia operacional total** entre las **ventas totales**. Es una medida *ponderada*, ya que las observaciones con mayores ventas influyen m√°s en el resultado. Responde a la pregunta: *‚Äú¬øQu√© porcentaje de todas las ventas se convirti√≥ en ganancia operacional?‚Äù*

- **Margen operacional promedio**: se calcula como el promedio de los m√°rgenes individuales de cada observaci√≥n. Aqu√≠ todas las filas tienen el mismo peso, sin importar el tama√±o de la venta. Responde a la pregunta: *‚ÄúEn promedio, ¬øqu√© porcentaje de cada transacci√≥n representa la ganancia operacional?‚Äù*

En conclusi√≥n, el margen total refleja la eficiencia global del retailer considerando el volumen de ventas, mientras que el margen promedio refleja la eficiencia promedio de cada operaci√≥n individual. Por ello, aunque ambos hablan de rentabilidad, los valores no necesariamente coinciden.

4. Teniendo en cuenta los resultados de arriba: 1. Tome el Retailer que vendio
mas 2. El estado que vendio mas 3.el producto que vendio mas y el metodo de ventas que vendio m√°s 4. Filtre y genere un dataframe con ellos 5. llamelo como un nuevo dataframe

```python
best_retailer = df.groupby("Retailer")["Total Sales"].sum().idxmax()
best_state = df.groupby("State")["Total Sales"].sum().idxmax()
best_product = df.groupby("Product")["Units Sold"].sum().idxmax()
best_method = df.groupby("Sales Method")["Total Sales"].sum().idxmax()

print("\nüèÜ Resultados")
print("Mejor Retailer:", best_retailer)
print("Mejor Estado:", best_state)
print("Producto m√°s vendido:", best_product)
print("Mejor m√©todo de venta:", best_method)

df_filtrado = df[
    (df["Retailer"] == best_retailer) &
    (df["State"] == best_state) &
    (df["Product"] == best_product) &
    (df["Sales Method"] == best_method)]

Best = df_filtrado

print("\nüìä DataFrame filtrado:")
display(Best)

print("Cantidad de filas encontradas:", len(Best))
```
