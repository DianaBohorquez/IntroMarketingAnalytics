1.2 Cuanto fue el margen promedio total ?

```python
margen_promedio = df["Operating Margin"].mean() * 100
print("El margen operativo promedio fue:", round(margen_promedio, 2), "%")
```

**1.3 Por que el margen promedio y no el total. Explique con palabras**



El **margen operativo** es un indicador expresado en forma de **proporción o porcentaje**.  
A diferencia de las ventas o de la ganancia, que son montos de dinero y por eso sí se pueden sumar, el margen muestra la **relación entre la ganancia y las ventas** de cada transacción.  

Si sumáramos todos los márgenes fila por fila, el resultado no tendría sentido, porque estaríamos acumulando porcentajes de distintas operaciones.  
Por eso, lo correcto es calcular el **promedio del margen**, que nos da un valor representativo de la eficiencia general de la empresa.  

En este caso, el margen promedio fue de aproximadamente **42%**, lo que significa que, en promedio, de cada dólar vendido, 42 centavos se convierten en ganancia operativa.



2. Cuales fueron las ventas totales por Retailer?

```python
ventas_por_retailer = df.groupby("Retailer")["Total Sales"].sum().sort_values(ascending=False)
print("Ventas totales por Retailer:")
print(ventas_por_retailer)
```

2.1 Cual fue la ganancia operacional total por retailer?

```python
ganancia_por_retailer = df.groupby("Retailer")[["Operating Profit"]].sum()
ganancia_por_retailer = ganancia_por_retailer.sort_values(by="Operating Profit", ascending=False)
print("Ganancia operacional total por Retailer:")
print(ganancia_por_retailer)
```

2.3 Tomando los dos resultados de arriba: El ganancia operacional total y las ventas totales divida ambos resultados. Le tiene que dar el margen total de cada Retailer.

```python
ventas_totales = df.groupby("Retailer")["Total Sales"].sum()
ganancia_total = df.groupby("Retailer")["Operating Profit"].sum()
margen_total = ganancia_total / ventas_totales
margen_total = margen_total * 100
print("📊 Margen Operacional Total por Retailer (%):")
print(margen_total.round(2))
```

2.4 Cual fue el margen operacional promedio por retailer?

```python
margen_promedio = df.groupby("Retailer")["Operating Margin"].mean() * 100
print("📊 Margen Operacional Promedio por Retailer (%):")
print(margen_promedio.round(2))
```

2.5 Por que los dos resultados anteriores le dieron diferente? Argumente

La diferencia se debe a la forma en que se calcula cada indicador:

- **Margen operacional total**: se obtiene dividiendo la **ganancia operacional total** entre las **ventas totales**. Es una medida *ponderada*, ya que las observaciones con mayores ventas influyen más en el resultado. Responde a la pregunta: *“¿Qué porcentaje de todas las ventas se convirtió en ganancia operacional?”*

- **Margen operacional promedio**: se calcula como el promedio de los márgenes individuales de cada observación. Aquí todas las filas tienen el mismo peso, sin importar el tamaño de la venta. Responde a la pregunta: *“En promedio, ¿qué porcentaje de cada transacción representa la ganancia operacional?”*

En conclusión, el margen total refleja la eficiencia global del retailer considerando el volumen de ventas, mientras que el margen promedio refleja la eficiencia promedio de cada operación individual. Por ello, aunque ambos hablan de rentabilidad, los valores no necesariamente coinciden.

3.1 Retailer que mas unidades totales vende? Saque cuanto vende en unidades cada retailer y busque el mayor

```python
unidades_por_retailer = df.groupby("Retailer")["Units Sold"].sum()
print("📦 Unidades totales vendidas por Retailer:")
print(unidades_por_retailer)
max_retailer = unidades_por_retailer.idxmax()
max_unidades = unidades_por_retailer.max()
print(f"\n👉 El retailer que más unidades vendió fue **{max_retailer}** con {max_unidades:,} unidades.")
```

4. Teniendo en cuenta los resultados de arriba: 1. Tome el Retailer que vendio
mas 2. El estado que vendio mas 3.el producto que vendio mas y el metodo de ventas que vendio más 4. Filtre y genere un dataframe con ellos 5. llamelo como un nuevo dataframe

```python
best_retailer = df.groupby("Retailer")["Total Sales"].sum().idxmax()
best_state = df.groupby("State")["Total Sales"].sum().idxmax()
best_product = df.groupby("Product")["Units Sold"].sum().idxmax()
best_method = df.groupby("Sales Method")["Total Sales"].sum().idxmax()

print("\n🏆 Resultados")
print("Mejor Retailer:", best_retailer)
print("Mejor Estado:", best_state)
print("Producto más vendido:", best_product)
print("Mejor método de venta:", best_method)

df_filtrado = df[
    (df["Retailer"] == best_retailer) &
    (df["State"] == best_state) &
    (df["Product"] == best_product) &
    (df["Sales Method"] == best_method)]

Best = df_filtrado

print("\n📊 DataFrame filtrado:")
display(Best)

print("Cantidad de filas encontradas:", len(Best))
```
