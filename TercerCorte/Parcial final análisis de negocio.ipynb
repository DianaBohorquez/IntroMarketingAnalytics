Si una columna no existe en una de las dos fuentes (por ejemplo, `views` solo para algunos posts), el impacto es significativo:

*   **Necesidad de Proxies**: Se podría recurrir a métricas proxy, pero esto introduce incertidumbre y reduce la fiabilidad de las conclusiones.
*   **Riesgo de Decisiones Subóptimas**: La falta de datos consistentes en métricas clave puede llevar a recomendaciones estratégicas incompletas o erróneas, subestimando o sobrestimando el desempeño real de una marca en esa área.

### 3.1. Preguntas de negocio: Datos faltantes

  'likesCount': 0.00%

  'videoViewCount': 60.50%

**Valores faltantes en Temu:**

**Análisis de los resultados:**

Las columnas `likesCount` y `commentsCount` están completamente llenas en ambos datasets. Sin embargo, `videoViewCount` presenta un desafío significativo, especialmente para Shein, donde el **60.50% de los posts no tienen este valor**. Temu tiene un porcentaje menor, con el 7.14%.

### Desde el punto de vista de negocio, ¿qué haríamos con esos posts?

La decisión sobre cómo manejar los valores faltantes depende del contexto y el objetivo del análisis:

1.  **Excluir (eliminar) posts con valores faltantes:**
    *   **Pros:** Simplicidad, garantiza trabajar con datos completos.
    *   **Contras:** Para Shein, eliminar el 60.50% de sus posts debido a la falta de `videoViewCount` significaría perder una gran cantidad de información y podría sesgar los resultados.
    *   **Recomendación:** No es recomendable para `videoViewCount` en Shein si el objetivo es comparar el engagement general de todos los posts.

2.  **Imputar (rellenar) los valores faltantes:**
    *   **Pros:** Permite mantener todos los posts en el análisis.
    *   **Contras:** Introducir valores artificiales puede distorsionar la distribución de los datos y llevar a conclusiones erróneas. Un cero podría ser incorrecto si la publicación es una imagen (donde `videoViewCount` no aplica).
    *   **Recomendación:** Extremadamente arriesgado para `videoViewCount`, especialmente en Shein, debido al alto volumen de datos faltantes.

3.  **Analizarlos aparte o crear categorías de análisis específicas:**
    *   **Pros:** Es la opción más robusta y que mejor respeta la integridad de los datos. Permite entender por qué faltan los datos y cómo eso impacta la interpretación.
    *   **Recomendación:** Esta es la mejor estrategia para `videoViewCount`:
        *   **Identificar la causa:** Es probable que `videoViewCount` sea nulo para posts que **no son videos** (ej. imágenes, carruseles). Su ausencia sería, entonces, esperable.
        *   **Segmentación del análisis:** Para calcular métricas de eficiencia como `engagement_rate`, lo ideal sería:
            *   Calcular el engagement para **videos** usando `(likes + comments) / videoViewCount`.
            *   Calcular el engagement para **imágenes/carruseles** usando `(likes + comments)` como métrica de interacción bruta, reconociendo que no es comparable directamente con los videos.

En resumen, para métricas con alto porcentaje de valores faltantes como `videoViewCount`, la mejor estrategia es **entender la razón de la falta de datos** y **segmentar el análisis** en función del tipo de contenido. Excluir indiscriminadamente o imputar con valores arbitrarios podría llevar a conclusiones muy engañosas.

```python
import numpy as np

def calculate_engagement_rate(df):
    # Convertir a numérico, forzando errores a NaN
    df['likesCount'] = pd.to_numeric(df['likesCount'], errors='coerce')
    df['commentsCount'] = pd.to_numeric(df['commentsCount'], errors='coerce')
    df['videoViewCount'] = pd.to_numeric(df['videoViewCount'], errors='coerce')
    df['videoPlayCount'] = pd.to_numeric(df['videoPlayCount'], errors='coerce')

    # Calcular interacciones totales
    df['total_interactions'] = df['likesCount'].fillna(0) + df['commentsCount'].fillna(0)

    # Definir 'effective_views' para el denominador
    # Usar videoViewCount si existe y es > 0
    # Si no, usar videoPlayCount si existe y es > 0
    # Si no, usar likesCount como proxy, asegurando que sea al menos 1 si hay interacciones.
    # Si no hay interacciones, establecer a 1 para que el engagement rate sea 0 y evitar div/0
    df['effective_views'] = df['videoViewCount'].fillna(df['videoPlayCount'])

    # Para los NaN restantes (principalmente imágenes sin 'views' directas), usar likesCount como proxy
    # Rellenar los NaNs en effective_views con likesCount. Si likesCount es 0, asegurar que el denominador sea al menos 1.
    df['effective_views'] = df['effective_views'].fillna(df['likesCount'])

    # Asegurar que effective_views sea al menos 1 si total_interactions > 0 para evitar div/0 para posts con interacciones pero 0 vistas/likes.
    # Si total_interactions es 0, effective_views puede ser 1, resultando en engagement_rate = 0.
    df['effective_views'] = np.where(
        df['effective_views'].fillna(0) == 0, # Si effective_views es 0 o NaN (y likesCount era 0)
        np.where(df['total_interactions'] > 0, 1, 1), # Si hay interacciones, pon 1. Si no hay interacciones, pon 1. (resultado: siempre 1 para 0 effective_views)
        df['effective_views'] # Usar el valor original si no es 0
    )

    # Calcular engagement_rate
    df['engagement_rate'] = df['total_interactions'] / df['effective_views']

    # Limpiar columnas temporales si no son necesarias
    df = df.drop(columns=['total_interactions', 'effective_views'])

    return df

# Aplicar la función a ambos dataframes
shein = calculate_engagement_rate(shein)
temu = calculate_engagement_rate(temu)

print("Primeros 5 registros de Shein con la nueva columna engagement_rate:")
display(shein[['likesCount', 'commentsCount', 'videoViewCount', 'videoPlayCount', 'engagement_rate']].head())

print("\nPrimeros 5 registros de Temu con la nueva columna engagement_rate:")
display(temu[['likesCount', 'commentsCount', 'videoViewCount', 'videoPlayCount', 'engagement_rate']].head())
```

### 4.1. Preguntas de negocio: Definición de eficiencia

**En tus palabras: ¿qué es “eficiencia” en una estrategia de contenido?**

La **“eficiencia” en una estrategia de contenido** se refiere a la capacidad de una marca para generar el máximo impacto (interacciones, atención, recordación de marca, e incluso conversiones) con la menor cantidad de recursos invertidos (tiempo, dinero, número de publicaciones, esfuerzo creativo). No se trata solo de producir una gran cantidad de contenido, sino de que cada pieza de contenido entregue el mayor retorno posible sobre la inversión realizada.

En otras palabras, una estrategia eficiente logra sus objetivos de comunicación y marketing optimizando el uso de sus activos y maximizando el engagement o la acción deseada por parte de la audiencia, en lugar de simplemente perseguir métricas de volumen.

**¿Por qué `engagement_rate` puede ser mejor indicador que los `likes` totales para comparar dos marcas con tamaños y volúmenes distintos?**

El `engagement_rate` es un indicador superior a los `likes` totales por las siguientes razones fundamentales, especialmente al comparar marcas de diferente escala:

1.  **Normalización por Alcance/Vistas:** El `engagement_rate` convierte una métrica absoluta (`likes` totales) en una métrica **relativa o de proporción** al dividir las interacciones por una medida del alcance o las vistas. Esto significa que no estamos comparando el número total de likes, sino la *efectividad* del contenido para captar la atención y generar interacción entre aquellos que lo vieron. Así, un post con pocas vistas pero muchas interacciones puede tener un `engagement_rate` más alto que un post con muchísimas vistas pero pocas interacciones.

2.  **Comparación Equitativa (Manzanas con Manzanas):** Cuando comparamos dos marcas como Shein y Temu, con volúmenes de publicación y audiencias (y por ende, alcance potencial) muy diferentes, los `likes` totales de la marca con más contenido (Shein) serán casi siempre mayores. Esto no implica necesariamente que Shein sea más eficiente, solo que tiene más oportunidades. El `engagement_rate`, al ser una proporción, permite comparar su capacidad de atraer interacción **por cada oportunidad de vista**, nivelando el campo de juego y ofreciendo una métrica más justa de la calidad del contenido.

3.  **Foco en la Calidad del Contenido y la Relevancia:** Un `engagement_rate` alto sugiere que el contenido es relevante, interesante o provocador para la audiencia que lo ve. Incluso si una marca tiene menos seguidores o publica menos, un `engagement_rate` elevado indica que su contenido resuena fuertemente con su público. Para un inversor, esto es crucial, ya que denota una conexión más profunda y una mayor capacidad de influencia por cada pieza de contenido producida.

4.  **Menos Susceptible a Métricas de Vanidad:** Los `likes` totales pueden ser inflados por tener una audiencia enorme pero pasiva, o por campañas de pago masivas que buscan alcance sin necesariamente profundizar la interacción. El `engagement_rate` mitiga este problema: si un contenido es visto por muchos pero no genera interacción significativa, su tasa de engagement será baja, revelando una ineficiencia en la activación de la audiencia.

```python
# Asegurarse de que las columnas 'timestamp' ya son datetime objects (ya se hizo en celdas previas)
# shein['timestamp'] = pd.to_datetime(shein['timestamp'])
# temu['timestamp'] = pd.to_datetime(temu['timestamp'])

# Crear las columnas 'day_of_week' y 'hour' para Shein
shein['day_of_week'] = shein['timestamp'].dt.day_name()
shein['hour'] = shein['timestamp'].dt.hour

# Crear las columnas 'day_of_week' y 'hour' para Temu
temu['day_of_week'] = temu['timestamp'].dt.day_name()
temu['hour'] = temu['timestamp'].dt.hour

print("Shein con 'day_of_week' y 'hour':")
display(shein[['timestamp', 'day_of_week', 'hour', 'engagement_rate']].head())

print("\nTemu con 'day_of_week' y 'hour':")
display(temu[['timestamp', 'day_of_week', 'hour', 'engagement_rate']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Crear una copia de los dataframes y añadir la columna 'brand'
shein_for_heatmap = shein[['day_of_week', 'hour', 'engagement_rate']].copy()
shein_for_heatmap['brand'] = 'Shein'

temu_for_heatmap = temu[['day_of_week', 'hour', 'engagement_rate']].copy()
temu_for_heatmap['brand'] = 'Temu'

# Concatenar los dataframes
combined_heatmap_df = pd.concat([shein_for_heatmap, temu_for_heatmap])

# Definir el orden de los días de la semana
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Calcular el engagement promedio por día de la semana y hora para Shein
shein_pivot = combined_heatmap_df[combined_heatmap_df['brand'] == 'Shein'].pivot_table(
    values='engagement_rate', index='day_of_week', columns='hour', aggfunc='mean'
).reindex(day_order)

# Calcular el engagement promedio por día de la semana y hora para Temu
temu_pivot = combined_heatmap_df[combined_heatmap_df['brand'] == 'Temu'].pivot_table(
    values='engagement_rate', index='day_of_week', columns='hour', aggfunc='mean'
).reindex(day_order)

# Crear los heatmaps
fig, axes = plt.subplots(1, 2, figsize=(20, 8), sharey=True)

sns.heatmap(shein_pivot, cmap='viridis', annot=True, fmt=".2f", linewidths=.5, ax=axes[0])
axes[0].set_title('Shein - Engagement Rate Promedio por Día y Hora')
axes[0].set_xlabel('Hora del Día')
axes[0].set_ylabel('Día de la Semana')

sns.heatmap(temu_pivot, cmap='viridis', annot=True, fmt=".2f", linewidths=.5, ax=axes[1])
axes[1].set_title('Temu - Engagement Rate Promedio por Día y Hora')
axes[1].set_xlabel('Hora del Día')
axes[1].set_ylabel('') # Para evitar etiquetas duplicadas en el eje y

plt.tight_layout()
plt.show()
```
